```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import random 

# Import API key
from config import api_key
#import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

#Create lists for data frame
total_temp = []
total_lat = []
total_lng = []
total_humidity = []
total_cloudiness = []
total_wind_speed = []
city_cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#cities
```

## Perform API Calls

```python
# OpenWeatherMap API Key
#api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

# Get weather data
query_url = url + "&q=" + city 
query_url
response = requests.get(query_url).json()
response
```

```python
for city in cities:
    query_url = url + "&q=" + city 
    print(query_url)
    try:
        response = requests.get(query_url).json() 
        city = response ["name"]
        temp = response["main"]["temp"]
        lat = response["coord"]["lat"]
        lng = response["coord"]["lon"]
        humidity = response["main"]["humidity"]
        cloudiness = response["clouds"]["all"]
        wind_speed = response["wind"]["speed"]
    
        total_temp.append(temp)
        city_cities.append(city)
        total_lat.append(lat)
        total_lng.append(lng)
        total_humidity.append(humidity)
        total_cloudiness.append(cloudiness)
        total_wind_speed.append(wind_speed)
    
    except:KeyError
    pass
```

```python
city_dict = {"City": city_cities, "Latitude": total_lat, "Longitude": total_lng, "Temperature": total_temp, "Humidity": total_humidity, "Cloudiness": total_cloudiness,"Wind Speed": total_wind_speed}

city_data = pd.DataFrame(city_dict)
city_data
```

```python
#Save CSV file for all the data retrieved 
city_data.to_csv("OM_WeatherPy.csv", index=False)
```
